```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# pd.set_option('display.max_rows', None)
```

```python
genshin_df= pd.read_csv('/content/Genshin_Impact_All_Character_Stat.csv.xls')
```

```python
genshin_df
```

```python
len(pd.unique(genshin_df.Character))    #no. of characters in the dataset
```

```python
sns.scatterplot(y=genshin_df['Base DEF'],x=genshin_df['Base ATK']);
```

```python
genshin_df["TDS"]=genshin_df['Base HP']**2/genshin_df['Base DEF']
```

```python
genshin_df
```

```python
lvl90_base_stats_df=genshin_df.loc[genshin_df.Lv==90,:]
lvl90_base_stats_df
```

```python
lvl90_base_stats_df['ratio']=lvl90_base_stats_df['Base DEF']/lvl90_base_stats_df['Base ATK']
lvl90_base_stats_df.sort_values(by='ratio', ascending=False);
lvl90_base_stats_df
```

```python
lvl90_base_stats_df["TDO"]=lvl90_base_stats_df["TDS"]*lvl90_base_stats_df["Base ATK"]
lvl90_base_stats_df
```

```python
lvl90_base_stats_df["attack efficiency"]=lvl90_base_stats_df["TDO"]*lvl90_base_stats_df["Base ATK"]**3;
lvl90_base_stats_df
```

```python
best_dps_df=lvl90_base_stats_df.sort_values(by=["attack efficiency"], ascending=False)
best_dps_df['DPS score']=best_dps_df['attack efficiency'].rank()
# best_dps_df=lvl90_base_stats_df.sort_values(by=["attack efficiency"], ascending=False).head(10)
best_dps_df.head(10)
```

```python
plt.figure(figsize=(16, 8))
plt.title('Best DPS')
sns.barplot(x='Character', y='attack efficiency', data=best_dps_df.head(10));
```

```python
worst_dps_df=lvl90_base_stats_df.sort_values(by=["attack efficiency"], ascending=False).tail(10)
worst_dps_df
```

```python
plt.figure(figsize=(16, 8))
plt.title('Worst DPS')
sns.barplot(x='Character', y='attack efficiency', data=worst_dps_df);
```

```python
best_survivor_df=lvl90_base_stats_df.sort_values(by=["TDS"], ascending=False)
best_survivor_df['Survivability score']=best_survivor_df['attack efficiency'].rank()
# best_survivor_df=lvl90_base_stats_df.sort_values(by=["TDS"], ascending=False).head(10)
best_survivor_df.head(10)
```

```python
plt.figure(figsize=(16, 8))
plt.title('Best Tanks')
sns.barplot(x='Character', y='TDS', data=best_survivor_df.head(10));
```

```python
worst_survivor_df=lvl90_base_stats_df.sort_values(by=["TDS"], ascending=False).tail(10)
worst_survivor_df
```

```python
plt.figure(figsize=(16, 8))
plt.title('Worst Tanks')
sns.barplot(x='Character', y='TDS', data=worst_survivor_df);
```

```python
most_consistent_damage_outputter_df=lvl90_base_stats_df.sort_values(by=["TDO"], ascending=False)
most_consistent_damage_outputter_df["Consistency score"]=most_consistent_damage_outputter_df['TDO'].rank()
# most_consistent_damage_outputter_df=lvl90_base_stats_df.sort_values(by=["TDO"], ascending=False)
most_consistent_damage_outputter_df.head(10)
```

```python
plt.figure(figsize=(16, 8))
plt.title('Most Consistent Damage Outputter')
sns.barplot(x='Character', y='TDO', data=most_consistent_damage_outputter_df);
```

```python
least_consistent_damage_outputter_df=lvl90_base_stats_df.sort_values(by=["TDO"], ascending=False).tail(10)
least_consistent_damage_outputter_df
```

```python
plt.figure(figsize=(16, 8))
plt.title('Least Consistent Damage Outputter')
sns.barplot(x='Character', y='TDO', data=least_consistent_damage_outputter_df);
```

```python
df1=pd.merge(lvl90_base_stats_df,best_dps_df[['Character','DPS score']],on='Character')
df1
```

```python
df2=pd.merge(df1,best_survivor_df[['Character','Survivability score']],on='Character')
df2
```

```python
df3=pd.merge(df2,most_consistent_damage_outputter_df[['Character','Consistency score']],on='Character')
df3
```

```python
df3["total rank"]=df3['DPS score']+df3['Survivability score']+df3['Consistency score']
```

```python
df4=df3.sort_values(by="total rank", ascending=False)
```

```python
df4['podium finish']=range(1, len(df4) + 1)
```
